## Hospital Patient Management Analysis With Pandas

### About This Notebook

This is a practice notebook for learning data cleaning, analysis, and visualization with Python.

### Dataset Notes

This notebook uses a small hospital dataset with patient information such as department, age, bill amount, admission date, and discharge status.

## 1. Setup

#### Import the Necessary Libraries for the Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

## 2. Create the Dataset

Build a small dataframe with realistic data.

```python
data = {
    'Patient_ID': [201, 202, 203, 204, 205, 206, 207, 208, 209, 210, 211, 212],
    'Name': ['Aarav', 'Ishita', 'Rohan', 'Fatima', np.nan, 'Vikram',
             'Ananya', 'Kabir', 'Zara', 'Dev', 'Meera', 'Arjun'],
    'Department': ['Cardiology', 'Orthopedics', 'Cardiology', 'Neurology', 'Orthopedics',
                   'Neurology', 'Cardiology', 'Orthopedics', 'Neurology', 'Cardiology',
                   np.nan, 'Orthopedics'],
    'Age': [45, 62, 38, 29, 71, 55, np.nan, 34, 60, 48, 27, 66],
    'Bill_Amount': [15000, 42000, 8000, np.nan, 30000, 18500,
                     9500, 27000, np.nan, 21000, 6000, 35000],
    'Admit_Date': ['2026-01-03', '2026-01-15', '2026-02-02', '2026-02-10', '2026-02-18',
                    '2026-03-01', '2026-03-05', '2026-03-12', '2026-03-20', '2026-04-02',
                    '2026-04-10', '2026-04-15'],
    'Discharged': [True, True, False, True, True, False, True, True, False, True, True, False]
}
df = pd.DataFrame(data)
df
```

### 3. Dataset Overview

Check info, missing values, and shape of the dataset.

```python
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nShape:", df.shape)
```

### 4. Handling Missing Names and Departments

Filling missing patient names with "Unknown", and missing departments with the most common department (mode).

```python
df['Name'] = df['Name'].fillna('Unknown')
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
df[['Name', 'Department']]
```

### 5. Filling Missing Ages

Filling missing ages with the median age of all patients.

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Age']
```

### 6. Filling Missing Bill Amounts by Department

Filling missing bill amounts using the average bill amount for each department.

```python
df['Bill_Amount'] = df['Bill_Amount'].fillna(df.groupby('Department')['Bill_Amount'].transform('mean'))
df['Bill_Amount']
```

### 7. Extracting Admit Month

Converting Admit_Date to a proper datetime type, then extracting the admission month.

```python
df["Admit_Date"] = pd.to_datetime(df["Admit_Date"])
df["Admit_Month"] = df["Admit_Date"].dt.month_name()
print(df["Admit_Date"].dtype)
df
```

### 8. Identifying Still-Admitted Patients

Creating a separate view of patients who are still admitted, sorted by bill amount (highest first).

```python
Still_Admitted = df[df['Discharged'] == False].sort_values(by='Bill_Amount', ascending=False)
print(Still_Admitted)
```

### 9. Average Bill by Department

Average bill amount for each department.

```python
Avg_bill = df.groupby('Department')['Bill_Amount'].mean()
print(Avg_bill)
```

### 10. Highest Revenue Department

Department that generated the most total revenue.

```python
High_Revenue = df.groupby('Department')['Bill_Amount'].sum().idxmax()
print(High_Revenue)
```

### 11. Age Group Segmentation

Splitting patients into age groups: young (0-39), middle_aged (40-60), and senior (60+).

```python
bins = [0, 39, 60, float("inf")]
labels = ['young', 'middle_aged', 'senior']
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)
print(df[['Age', 'Age_group']].head(5))
```

### 12. Pivot Table

Pivot table showing each department's total patients vs discharged/not-discharged split.

```python
Pivot = df.pivot_table(index='Department', columns='Discharged', values='Patient_ID', aggfunc='count', margins=True)
print(Pivot)
```

### 13. Data Table Visualization

Rendering the full hospital dataframe as a matplotlib table.

```python
fig, ax = plt.subplots(figsize=(14, 6))
ax.axis("off")

table = ax.table(
    cellText=df.values,
    colLabels=df.columns,
    cellLoc="center",
    loc="center"
)

table.auto_set_font_size(False)
table.set_fontsize(8)
table.scale(1.2, 1.2)

plt.tight_layout()
plt.show()
```

### 14. Department Revenue Bar Chart

Total bill amount collected by each department.

```python
plt.figure(figsize=(10, 5))

dept_bill = df.groupby('Department')['Bill_Amount'].sum().sort_values(ascending=False)

plt.bar(
    dept_bill.index,
    dept_bill.values,
    color='royalblue',
    alpha=0.7,
    edgecolor='navy'
)

plt.title('Total Bill Amount by Department', fontsize=14, fontweight='bold')
plt.xlabel('Department')
plt.ylabel('Total Bill Amount')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### 15. Key Takeaways

- Cardiology and Orthopedics had the highest patient volume in the dataset.
- Missing values were handled carefully (name, department, age, bill amount) to keep the analysis reliable.
- Creating new columns such as Admit_Month and Age_group helped reveal useful patterns.
- Clear column names and grouped summaries made the hospital data easier to understand.
